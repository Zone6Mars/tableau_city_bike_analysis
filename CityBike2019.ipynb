```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
# Files to load
oct2019 = "Data/2019/JC-201910-citibike-tripdata.csv"
nov2019 = "Data/2019/JC-201911-citibike-tripdata.csv"
dec2019 = "Data/2019/JC-201912-citibike-tripdata.csv"
```

```python
oct2019_df = pd.read_csv(oct2019)
nov2019_df = pd.read_csv(nov2019)
dec2019_df = pd.read_csv(dec2019)
```

```python
# Combine All Three October DataFrames from 2018-2020 Into Single DataFrame
octnov_df = oct2019_df.append([nov2019_df], ignore_index=True)
octnov_df.head()
```

```python
octnovdec_df = octnov_df.append([dec2019_df], ignore_index=True)
octnovdec_df.head()
```

```python
BikeData_2019_df = octnovdec_df
BikeData_2019_df.head()
```

```python
# Convert All DateTimes to "%Y-%m-%d %H:%M:%S" Format
BikeData_2019_df["starttime"] = pd.to_datetime(BikeData_2019_df["starttime"])
BikeData_2019_df["stoptime"] = pd.to_datetime(BikeData_2019_df["stoptime"])
```

```python
Clean_BikeData_2019_df = BikeData_2019_df.drop(columns=['birth year', 'gender', 'bikeid','start station id','end station id'])
```

```python
Clean_BikeData_2019_df['usertype'].replace('Subscriber','member',inplace = True)
```

```python
Clean_BikeData_2019_df['usertype'].replace('Customer','casual',inplace = True)
```
