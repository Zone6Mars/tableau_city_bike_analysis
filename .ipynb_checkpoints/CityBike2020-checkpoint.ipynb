```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
# Files to load
jan2020 = "Data/2020/JC-202001-citibike-tripdata.csv"
feb2020 = "Data/2020/JC-202002-citibike-tripdata.csv"
mar2020 = "Data/2020/JC-202003-citibike-tripdata.csv"
apr2020 = "Data/2020/JC-202004-citibike-tripdata.csv"
may2020 = "Data/2020/JC-202005-citibike-tripdata.csv"
jun2020 = "Data/2020/JC-202006-citibike-tripdata.csv"
jul2020 = "Data/2020/JC-202007-citibike-tripdata.csv"
aug2020 = "Data/2020/JC-202008-citibike-tripdata.csv"
sep2020 = "Data/2020/JC-202009-citibike-tripdata.csv"
oct2020 = "Data/2020/JC-202010-citibike-tripdata.csv"
nov2020 = "Data/2020/JC-202011-citibike-tripdata.csv"
dec2020 = "Data/2020/JC-202012-citibike-tripdata.csv"
```

```python
jan2020_df = pd.read_csv(jan2020)
feb2020_df = pd.read_csv(feb2020)
mar2020_df = pd.read_csv(mar2020)
apr2020_df = pd.read_csv(apr2020)
may2020_df = pd.read_csv(may2020)
jun2020_df = pd.read_csv(jun2020)
jul2020_df = pd.read_csv(jul2020)
aug2020_df = pd.read_csv(aug2020)
sep2020_df = pd.read_csv(sep2020)
oct2020_df = pd.read_csv(oct2020)
nov2020_df = pd.read_csv(nov2020)
dec2020_df = pd.read_csv(dec2020)
```

```python
janthrudec_df = jan2020_df.append([feb2020_df, mar2020_df, apr2020_df, may2020_df, jun2020_df, jul2020_df, aug2020_df, sep2020_df, oct2020_df, nov2020_df, dec2020_df], ignore_index=True)
janthrudec_df.head()
                    
```

```python
janthrudec_df.count()
```

```python
BikeData_2020_df = janthrudec_df
BikeData_2020_df.head()
```

```python
# Convert All DateTimes to "%Y-%m-%d %H:%M:%S" Format
BikeData_2020_df["starttime"] = pd.to_datetime(BikeData_2020_df["starttime"])
BikeData_2020_df["stoptime"] = pd.to_datetime(BikeData_2020_df["stoptime"])
```

```python
Clean_BikeData_2020_df = BikeData_2020_df.drop(columns=['birth year', 'gender', 'bikeid'])
```
